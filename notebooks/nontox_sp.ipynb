```python
import pandas as pd
from Bio import SeqIO
```

```python
nontox_dict = {}
for record in SeqIO.parse("../data/nontox.fasta", "fasta"):
    identifier = record.id[1:].split("|")[1]
    seq = record.seq
    nontox_dict[identifier] = seq
nontox = pd.DataFrame().from_dict(nontox_dict, orient="index", columns=["identifier", "seq"])
```
